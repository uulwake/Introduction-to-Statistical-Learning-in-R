## 5. Visualisasikan Full Data

Seperti yang sudah kita kerjakan di nomor 8, kita akan menggunakan PairGrid. Penjelasan kenapa kita lebih memilik PairGrid dibandingkan dengan pairplot sudah dijelaskan di sana

```python
g = sns.PairGrid(auto, size=10)
g.map_upper(plt.scatter, s=3)
g.map_diag(plt.hist)
g.map_lower(plt.scatter, s=3)
g.fig.set_size_inches(12, 12)
```
