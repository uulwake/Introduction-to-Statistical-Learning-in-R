#### B. Pairplot

Kita menggunakan seaborn untuk pairplot. Artikel yang bagus mengenai pairplot dapat di baca [di sini](https://towardsdatascience.com/visualizing-data-with-pair-plots-in-python-f228cf529166)
> The default pairs plot in seaborn only plots numerical columns

Secara default, pairplot menggunakan python seaborn hanya dapat memplot numerical column (kolom yang mengandung angka angka) sedangkan categorical column (yang isinya hanya yes or no atau sejenisnya) tidak dapat di tampilkan.


````python
sns.pairplot(college)
````

Dari gambar di atas, dapat dilihat ada dua grafik yaitu histogram dan scatterplot. Histogram merupakan plot dengan satu variabel sementara itu scatterplot merupakan plot yang menggunakan dua buah variabel

Jika gambarnya di perbesar, akan terlihat bahwa kita tidak dapat memplot categorical data. Hal ini bisa diakali dengan cara menggunakan atribut `hue`

Selain itu kita juga dapat menggunakan seaborn PairGrid method yang dapat dilihat lebih lanjut [di sini](https://seaborn.pydata.org/generated/seaborn.PairGrid.html).

Karena di soal meminta 10 kolom pertama untuk di plot, maka kita menggunakan parameter `vars`. 


Informasi mengenai pandas.DataFrame.iloc dapat dilihat [di sini](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.iloc.html)
> Purely integer-location based indexing for selection by position.

````python
g = sns.PairGrid(college, vars=college.iloc[:,1:11], hue='Private')
g.map_upper(plt.scatter, s=3)
g.map_diag(plt.hist)
g.map_lower(plt.scatter, s=3)
g.fig.set_size_inches(12, 12)
````

Sedikit penjelasan mengenai perbedaan `seaborn.pairplot` dan `seaborn.PairGrid`.<br> 
`seaborn.pairplot` merupakan high level dari `seaborn.PairGrid` sehingga `seaborn.PairGrid` lebih direkomendasikan jika kita menginginkan fleksibilitas.
> *This is a high-level interface for PairGrid that is intended to make it easy to draw a few common styles. You should use PairGrid directly if you need more flexibility.*

### D. Elite qualitative variable

Kita ingin membuat sebuah variabel kualitatif (nilainya diskrit) yang baru<br>
Penjelasan mengenai `pandas.DataFrame.loc` [di sini](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.loc.html) <br>
Penjelasan mengenai `pandas.DataFrame.fillna` [di sini](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.loc.html)


````python
college.loc[college['Top10perc']>50, 'Elite'] = 'Yes'
college['Elite'] = college['Elite'].fillna('No')
# plot Elite vs Outstate
sns.boxplot(x='Elite', y='Outstate', data=college)
````

### E. Histogram untuk Variabel Kualitatif

Di sini kita menggunakan `pandas.cut` untuk membuat lebar dari histogramnya.<br>
Dokumen `pandas.cut` bisa dilihat [di sini](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.cut.html)

````python
college.head()
````

````python
# membuat bin dari histogramnya
college['Enroll'] = pd.cut(college['Enroll'], 5, labels=['Sangat Rendah', 'Rendah', 'Normal', 'Tinggi', 'Sangat Tinggi'])
college['Grad.Rate'] = pd.cut(college['Grad.Rate'], 7, labels=['Bahaya','Sangat Rendah', 'Rendah', 'Normal', 'Tinggi', 'Sangat Tinggi', 'Luar Biasa'])
````

````python
# plot histogramnya
fig = plt.figure()
plt.subplot(111)
college['Enroll'].value_counts().plot(kind='bar', title='Enroll')
````

````python
plt.subplot(111)
college['Grad.Rate'].value_counts().plot(kind='bar', title='Grad.Rate');
````

### F. Analisi dari data

Anda dapat melakukan analisis sendiri. Tingal melihat hasil pairplot. Lihat yang mana memiliki korelasi tinggi (positif maupun negatif) atapun yang sama sekali tidak memiliki korelasi.Contohnya ialah Acceptance vs F.Undergrad memiliki korelasi positif yang kuat.

````python
plt.plot(college['F.Undergrad'], college['Accept'], 'ro')
plt.show()
````

````python
# hitung korelasi antara F.Undergrad dan Accept
np.corrcoef(college['F.Undergrad'], college['Accept'])[0, 1]
````

Dari hasil di atas diketahui bahwa korelasi antara F.Undergrad dan Accept cukup tinggi.<br>
Anda bisa mencari analisis yang lain.<br>
Penjelasan lebih lanjut tentang `numpy.corrcoef` bisa di lihat [di sini](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.corrcoef.html)
